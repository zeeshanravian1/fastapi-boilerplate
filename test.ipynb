```python
# "Duplicate entry 'admin' for key 'role.role_name'"
# i want to extract first '' values and last '' values by which i can split into . and then create a message
# like "role_name=admin already exists"

import re

def extract_values(message):
    values = re.findall(r"'(.*?)'", message)
    return values

def create_message(values):
    return f"{values[0]}={values[1]} already exists"

def main():
    message = "Duplicate entry 'admin' for key 'role.role_name'"
    values = extract_values(message)
    print(create_message(values))

main()
```

```python
import re

# Extract values from the message
values = re.findall(
    r"'(.*?)'", "Duplicate entry 'admin' for key 'role.role_name'"
)

# Extract the column name from the table.column format
column_name = values[1].split(".")[-1]

# Create the error message
error_message = f"{values[0]}={column_name} already exists"

print(error_message)
```
